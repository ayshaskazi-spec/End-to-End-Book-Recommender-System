```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('BX-Books.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
books.shape
```

```python
books = books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-L']]
```

```python
books.rename(columns={"Book-Title":'title', 'Book-Author':'author', "Year-Of-Publication":'year',"Publisher":"publisher","Image-URL-L":"image-url"}, inplace=True)
```

```python
books.head()
```

```python
users = pd.read_csv('BX-Users.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
users.shape
```

```python
books.rename(columns={"User-ID":'user_id', 'Location':'location', "Age":'age'}, inplace=True)
```

```python
users.head()
```

```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
ratings.shape
```

```python
ratings.rename(columns={"User-ID":'user_id', 'Book-Rating':'rating'}, inplace=True)
```

```python
ratings.head()
```

```python
print(books.shape, users.shape, ratings.shape, sep='\n')
```

```python
ratings['user_id'].value_counts()
```

```python
#Storing users who had at least rated more than 200 books
x = ratings['user_id'].value_counts()>200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
#Merging ratings and books
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'num_of_rating'}, inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
```

```python
final_rating.head()
```

```python
#Taking the books which got at least 50 ratings

final_rating = final_rating[final_rating['num_of_rating']<=50]
```

```python
final_rating.head()
```

```python
#drop the duplicates
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
final_rating['author'] = final_rating['author'].fillna('Unknown')
final_rating['publisher'] = final_rating['publisher'].fillna('Unknown')
final_rating['image-url'] = final_rating['image-url'].fillna('https://via.placeholder.com/150')
```

```python
final_rating.isnull().sum()
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```
